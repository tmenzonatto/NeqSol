# Summary

**Fiz 3 modelos:**
- *No MP* = mesmos descritores do original, sem o ponto de fusão
- *No MP + 25 RDKit* = adicionei 25 descritores do RDKit (selecionados com SelectKBest)
- *No MP + 4 RDKit* = adicionei os 4 melhores do RDKit (selecionei os que apresentaram maior importância no modelo anterior)

**Resultados**

| Method | R2 | RMSE | %LogS +/- 0.7 | %LogS +/- 1.0 |
|:----|:---:|:---:|:---:|:---:|
| Original (CV)               | 0.87 | 0.88 | 66.3 | 79.4 |
| Original (Test set)         | 0.93 | 0.71 | 66.3 | 84.2 |
| No MP (CV)                  | 0.84 | 0.97 | 62.8 | 77.3 |
| No MP (Test set)            | 0.92 | 0.72 | 66.3 | 84.2 |
| No MP + 25 RDKit (CV)       | 0.87 | 0.86 | 67.2 | 80.6 |
| No MP + 25 RDKit (Test set) | 0.92 | 0.71 | 76.8 | 87.4 |
| No MP + 4 RDKit (CV)        | 0.87 | 0.86 | 68.2 | 80.1 |
| No MP + 4 RDKit (Test set)  | 0.92 | 0.72 | 74.7 | 86.3 |


**Conclusão**

Os modelos sem ponto de fusão acrescentando os descritores do RDKit ficaram bem próximos do original. Podemos optar por adicionar só esses 4 descritores para deixar o modelo mais simples.

# Define functions

```python
# Number of repeats (No artigo os resultados apresentados são a média de 100 runs do modelo)
# Como ia demorar muito, fiz 20
n_rep = 100
```

```python
# Metrics
def rmse(y_true, y_pred):
    return metrics.mean_squared_error(y_true, y_pred, squared=False)


def within_range07(y_true, y_pred):
    logS_range = 0.7
    a = y_true - logS_range
    b = y_true + logS_range
    c = np.array((y_pred >= a)  &  (y_pred <= b))

    return float("{:.2f}".format((c.sum()/len(y_pred))*100))


def within_range10(y_true, y_pred):
    logS_range = 1.0
    a = y_true - logS_range
    b = y_true + logS_range
    c = np.array((y_pred >= a)  &  (y_pred <= b))

    return float("{:.2f}".format((c.sum()/len(y_pred))*100))
```

```python
# Define cross-validation object
cv = KFold(n_splits=5, shuffle=True, random_state=0)

# Define scoring metrics
scoring = {'R2': 'r2',
           'RMSE': metrics.make_scorer(rmse),
           'within_range07': metrics.make_scorer(within_range07),
           'within_range10': metrics.make_scorer(within_range10)}
```

```python
# Function to print model metrics
def print_metrics(model, X_test, y_test):
    print('R2                ', round(metrics.r2_score(y_test, model.predict(X_test)), 4))
    print('RMSE              ', round(rmse(y_test, model.predict(X_test)), 4))
    print('within_range (0.7)', round(within_range07(y_test, model.predict(X_test)), 4))
    print('within_range (1.0)', round(within_range10(y_test, model.predict(X_test)), 4))
```

```python
# Boxplot metrics
def boxplot_metrics(df):
    fig, ax = plt.subplots(1, 4, figsize=(22, 5))
    df.boxplot(column='mean_R2', ax=ax[0], fontsize=14)
    df.boxplot(column='mean_RMSE', ax=ax[1], fontsize=14)
    df.boxplot(column='mean_wr07', ax=ax[2], fontsize=14)
    df.boxplot(column='mean_wr10', ax=ax[3], fontsize=14)
    plt.show()
```

```python
# Permutation importance
def pi_plot(model, X, y, title):
    pi = permutation_importance(model, X, y, n_repeats=20,
                                random_state=0,
                                scoring='neg_root_mean_squared_error')

    plt.figure(figsize=(14,10))
    plt.barh(X.columns, pi.importances_mean, xerr=1.96*pi.importances_std)
    plt.title(title)
    plt.show()
```

```python
# Plot predictions x experimental values for the best model
def plot_preds(model, X_test, y_test):
    # Style
    sns.set(color_codes=True)
    sns.set_style("white")
    # Predict
    y_pred = model.predict(X_test)
    # Plot
    ax = sns.regplot(x=y_test, y=y_pred, scatter_kws={'alpha':0.4})
    ax.set_xlabel('Experimental LogS', fontsize='large', fontweight='bold')
    ax.set_ylabel('Predicted LogS', fontsize='large', fontweight='bold')
    ax.figure.set_size_inches(7, 7)
    plt.show()
```

```python
# Remove highly correlated descriptors
from itertools import combinations

remove_descs = []
correlates_with = []

corr_matrix = X_train2.corr()
comb = combinations(corr_matrix.columns, 2)

for c1, c2 in comb:
    if (abs(corr_matrix.loc[c1, c2]) > 0.9) & (c1 not in remove_descs):
        remove_descs.append(c2)
        correlates_with.append(c1)
```
