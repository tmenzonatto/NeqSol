# Define functions

```python
# Number of repeats (No artigo os resultados apresentados são a média de 100 runs do modelo)
# Como ia demorar muito, fiz 20
n_rep = 100
```

```python
# Metrics
def rmse(y_true, y_pred):
    return metrics.mean_squared_error(y_true, y_pred, squared=False)


def within_range07(y_true, y_pred):
    logS_range = 0.7
    a = y_true - logS_range
    b = y_true + logS_range
    c = np.array((y_pred >= a)  &  (y_pred <= b))

    return float("{:.2f}".format((c.sum()/len(y_pred))*100))


def within_range10(y_true, y_pred):
    logS_range = 1.0
    a = y_true - logS_range
    b = y_true + logS_range
    c = np.array((y_pred >= a)  &  (y_pred <= b))

    return float("{:.2f}".format((c.sum()/len(y_pred))*100))
```

```python
# Define cross-validation object
cv = KFold(n_splits=10, shuffle=True, random_state=0)

# Define scoring metrics
scoring = {'R2': 'r2',
           'RMSE': metrics.make_scorer(rmse),
           'within_range07': metrics.make_scorer(within_range07),
           'within_range10': metrics.make_scorer(within_range10)}
```

```python
# Function to print model metrics
def print_metrics(model, X_test, y_test):
    print('R2                ', round(metrics.r2_score(y_test, model.predict(X_test)), 4))
    print('RMSE              ', round(rmse(y_test, model.predict(X_test)), 4))
    print('within_range (0.7)', round(within_range07(y_test, model.predict(X_test)), 4))
    print('within_range (1.0)', round(within_range10(y_test, model.predict(X_test)), 4))
```

```python
# Boxplot metrics
def boxplot_metrics(df):
    fig, ax = plt.subplots(1, 4, figsize=(22, 5))
    df.boxplot(column='mean_R2', ax=ax[0], fontsize=14)
    df.boxplot(column='mean_RMSE', ax=ax[1], fontsize=14)
    df.boxplot(column='mean_wr07', ax=ax[2], fontsize=14)
    df.boxplot(column='mean_wr10', ax=ax[3], fontsize=14)
    plt.show()
```

```python
# Permutation importance
def pi_plot(model, X, y, title):
    pi = permutation_importance(model, X, y, n_repeats=20,
                                random_state=0,
                                scoring='neg_root_mean_squared_error')

    plt.figure(figsize=(14,10))
    plt.barh(X.columns, pi.importances_mean, xerr=1.96*pi.importances_std)
    plt.title(title)
    plt.show()
```

```python
# Plot predictions x experimental values for the best model
def plot_preds(model, X_test, y_test):
    # Style
    sns.set(color_codes=True)
    sns.set_style("white")
    # Predict
    y_pred = model.predict(X_test)
    # Plot
    ax = sns.regplot(x=y_test, y=y_pred, scatter_kws={'alpha':0.4})
    ax.set_xlabel('Experimental LogS', fontsize='large', fontweight='bold')
    ax.set_ylabel('Predicted LogS', fontsize='large', fontweight='bold')
    ax.figure.set_size_inches(7, 7)
    plt.show()
```

# Import dataset and separate train/test sets

```python
# Import dataset
df = pd.read_csv(r'DATA/Descriptors/water_set_wide_descriptors.csv')
```

```python
df["Train_test"].value_counts()
```

```python
df_train = df[df["Train_test"] == "Train"]
df_test = df[df["Train_test"] == "Test"]
```

```python
# Descriptors without MP
descs = ['MW', 'Volume', 'G_solv', 'DeltaG_sol', 'solv_dip',
         'LsoluHsolv', 'LsolvHsolu', 'SASA', 'O_charges', 'C_charges',
         'Most_neg', 'Most_pos', 'Het_charges']

X_train = df_train[descs]
y_train = df_train['LogS']

X_test = df_test[descs]
y_test = df_test['LogS']
```

# Retrain model without melting points

```python
# Train 20 models with 500 trees, plot metrics
metrics_df = []

for i in range(n_rep):
    random_state = random.randint(0,999999)
    reg = make_pipeline(preprocessing.StandardScaler(),
                        ExtraTreesRegressor(n_estimators=500, random_state=random_state))
    scores = cross_validate(reg, X_train, y_train, cv=cv, scoring=scoring)
    metrics_dict = {'model':i,
                    'random_state':random_state,
                    'mean_R2':scores['test_R2'].mean(),
                    'mean_RMSE':scores['test_RMSE'].mean(),
                    'mean_wr07':scores['test_within_range07'].mean(),
                    'mean_wr10':scores['test_within_range10'].mean()}

    metrics_df.append(metrics_dict)
```

## Metrics

```python
# Best model - highest R2
metrics_df = pd.DataFrame(metrics_df)
metrics_df.sort_values(by=['mean_R2'], ascending=False)
```

```python
# Best model - lowest RMSE
metrics_df = pd.DataFrame(metrics_df)
metrics_df.sort_values(by=['mean_RMSE'], ascending=True)
```

```python
# Boxplot metrics
boxplot_metrics(metrics_df)
```

```python
# Fit best model, predict on test set
ET_no_MP = ExtraTreesRegressor(n_estimators=500, random_state=1459)
ET_no_MP.fit(X_train, y_train)
```

```python
print_metrics(ET_no_MP, X_test, y_test)
```

```python
# permutation_importance
pi_plot(ET_no_MP, X_test, y_test, "Test permutation importance. Model: 13 descriptors (MP removed)")
```

```python
# Plot predictions x experimental values for the best model
plot_preds(ET_no_MP, X_test, y_test)
```

# Retrain model with RDKit descriptors


```python
# RDKit descriptors
names = [name[0] for name in Descriptors.descList]
calc = MoleculeDescriptors.MolecularDescriptorCalculator(names)
rdkit_descs = [calc.CalcDescriptors(Chem.MolFromSmiles(cpd)) for cpd in df["SMILES"]]
descriptors_df = pd.DataFrame(rdkit_descs, columns=names)

# Add descriptors to original df
df[names] = descriptors_df

# Consider 13 original descriptors + RDKit's 208
all_descs = descs + names

# Recreate train/test sets - y sets are the same
df_train = df[df["Train_test"] == "Train"]
df_test = df[df["Train_test"] == "Test"]

X_train2 = df_train[all_descs]
X_test2 = df_test[all_descs]
```

```python
# Remove highly correlated descriptors
from itertools import combinations

remove_descs = []
correlates_with = []

corr_matrix = X_train2.corr()
comb = combinations(corr_matrix.columns, 2)

for c1, c2 in comb:
    # if a variable is already marked to be removed, do not check correlations to it
    if (abs(corr_matrix.loc[c1, c2]) > 0.9) and (c1 not in remove_descs) and (c2 not in remove_descs):
        remove_descs.append(c2)
        correlates_with.append(c1)
```

```python
pd.set_option('display.max_rows', None)
```

```python
d = {"Removed descriptors": remove_descs, "Correlated with": correlates_with}
pd.DataFrame(d)
```

Note: Descriptor "SASA" should be removed, correlates with "Volume".

```python
print(f"Number of correlated descriptors: {len(set(remove_descs))}")
```

```python
all_descs_no_corr = [desc for desc in all_descs if desc not in remove_descs]
print(f"Number of non-correlated descriptors: {len(set(all_descs_no_corr))}")
```

```python
X_train2_no_corr = X_train2[all_descs_no_corr]
X_test2_no_corr = X_test2[all_descs_no_corr]
```

There are still too many descriptors. We will select the top 30 using SelectKBest.

```python
# Select features
from sklearn.feature_selection import SelectKBest, f_regression

# Define feature selection
fs = SelectKBest(score_func=f_regression, k=30)

# Apply feature selection
X_selected = fs.fit_transform(X_train2_no_corr, y_train)
```

```python
selected_cols = list(fs.get_feature_names_out(input_features=X_train2_no_corr.columns))
selected_cols
```

Interessante, nem todas as colunas do modelo original foram selecionadas.

Vou manter as do modelo e os novos descritores sugeridos

```python
final_descs = descs + [val for val in selected_cols if val not in descs]
final_descs
```

25 descritores 1D/2D adicionados

```python
# Final train set
X_train_with_RDKit = X_train2[final_descs]
X_test_with_RDKit = X_test2[final_descs]
```

```python
X_train_with_RDKit.shape
```

```python
# Train 20 models with 500 Trees and 38 descriptors
metrics_df2 = []

for i in range(n_rep):
    random_state = random.randint(0,999999)
    reg = make_pipeline(preprocessing.StandardScaler(),
                        ExtraTreesRegressor(n_estimators=500, random_state=random_state))
    scores = cross_validate(reg, X_train_with_RDKit, y_train, cv=cv, scoring=scoring)
    metrics_dict = {'model':i,
                    'random_state':random_state,
                    'mean_R2':scores['test_R2'].mean(),
                    'mean_RMSE':scores['test_RMSE'].mean(),
                    'mean_wr07':scores['test_within_range07'].mean(),
                    'mean_wr10':scores['test_within_range10'].mean()}

    metrics_df2.append(metrics_dict)
```

## Metrics

```python
# Best model - highest R2
metrics_df2 = pd.DataFrame(metrics_df2)
metrics_df2.sort_values(by=['mean_R2'], ascending=False)
```

```python
# Best model - lowest RMSE
metrics_df2 = pd.DataFrame(metrics_df2)
metrics_df2.sort_values(by=['mean_RMSE'], ascending=True)
```
